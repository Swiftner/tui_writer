## ğŸ¤ Voice-to-AI

Complete interactive voice assistant that:
- Records audio from your microphone
- Transcribes speech using Whisper (via LiveTranscriber)
- Sends transcript to AI with web search
- Returns spoken response

**Usage:** Press Enter to record, type 'q' to quit.

```python
#| eval: false

from lisette import Chat
from dotenv import load_dotenv
from tui_writer.live import LiveTranscriber
import sounddevice as sd

load_dotenv()

print("âš™ï¸  Initializing AI...")
chat = Chat("gemini/gemini-2.5-flash", search="l")
print("âœ… AI ready!")

print("âš™ï¸  Loading Whisper model...")
transcriber = LiveTranscriber(model_id="tiny", language="en")
print("âœ… Whisper ready!")

print("\nğŸ¤ Voice-to-AI Assistant")

def record_audio(duration=5, sample_rate=16000):
    print(f"\nğŸ™ï¸  Recording {duration}s... Speak now!")
    audio = sd.rec(int(duration * sample_rate), samplerate=sample_rate, channels=1, dtype='float32')
    sd.wait()
    print("âœ… Recording complete!")
    return audio.flatten(), sample_rate

def transcribe_audio(audio_array, sample_rate):
    """Transcribe audio using LiveTranscriber's Whisper model."""
    print("ğŸ”„ Processing audio...")
    segments, _ = transcriber.transcribe_model.transcribe(
        audio_array,
        language=transcriber.language,
        beam_size=1,
        condition_on_previous_text=False,
    )
    transcript = " ".join([seg.text.strip() for seg in segments]).strip()
    return transcript

# Main loop
while True:
    cmd = input("\n Write 'q' to quit, or click enter to record: ").strip().lower()
    
    if cmd == 'q':
        print("ğŸ‘‹ Goodbye!")
        break
    
    # Record and transcribe
    audio_array, sample_rate = record_audio(duration=5)
    transcript = transcribe_audio(audio_array, sample_rate)
    
    if transcript:
        print(f"âœ… Transcription complete!")
        print(f"ğŸ¤ You: {transcript}")
        
        # Get AI response
        print("ğŸ¤– AI thinking...")
        response = chat(transcript)
        ai_text = response.choices[0].message.content
        
        print(f"\nğŸ¤– AI: {ai_text}")
    else:
        print("âš ï¸  No speech detected. Try speaking louder or closer to the mic.")
```
